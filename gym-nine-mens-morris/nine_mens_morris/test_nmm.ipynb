```python
from nine_mens_morris import nmm
import numpy as np
```

```python
env = nmm.env(render_mode='human')
env.reset()
env.max_steps = 100
while not (env.terminations['player_0'] or env.truncations['player_0']):
    o =  env.observe(env.agent_selection)
    mask = o['action_mask']
    o = o['observation']
    print(mask.tolist())
    # print(o)
    action = mask.nonzero()[0][np.random.choice(int(sum(mask)))]
    print(f'Action: {action}, count: {env.step_}, max_steps={env.max_steps}')
    env.step(action)
    print(env.rewards)
    # print(env.infos)
    env.render()

env.terminations['player_0'], env.truncations['player_0']
```
