```python
from nmm.env import NMMEnv
import numpy as np
```

```python
env = NMMEnv(render_mode='human')
env.reset()
count = 0
while not env.terminations['player_0']:
    o =  env.observe(env.agent_selection)
    mask = o['action_mask']
    o = o['observation']
    print(mask.tolist())
    # print(o)
    action = mask.nonzero()[0][np.random.choice(int(sum(mask)))]
    print(f'Action: {action}, count: {count}')
    env.step(action)
    print(env.rewards)
    # print(env.infos)
    env.render()

    count += 1
    if count > 1000:
        break
```
